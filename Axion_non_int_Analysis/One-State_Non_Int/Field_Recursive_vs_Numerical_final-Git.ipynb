## Whitaker Function ( also will be used to fit to Far field)

```python
from mpmath import*
def func_W_s_far(r,a,b): # l is equal to 1 radial laplace solution

    mu=0
    l=1
    
    gamma_c=0
    beta= 1.7526648513+ 0.703934*gamma_c-0.109101*gamma_c**2+0.013436*gamma_c**3+0.017778*gamma_c**4-0.018281*gamma_c**5 +0.005129*gamma_c**6
    alpha=3.4951309897- 0.117682*gamma_c-0.391600*gamma_c**2+0.191882*gamma_c**3-0.041828*gamma_c**4-0.041507*gamma_c**5 +0.033020*gamma_c**6
    W=[]
    k=beta*(1+mu)**(-1/2)
    for i in range(0,len(r)):
        z=r[i]
        w_s0=b*(1/(l*(l+1)))*alpha*(1/(2**(beta)*z))* whitw(beta, -1/2, 2*z)
        w_s=( a - b*(1/(l*(l+1)))*alpha)*(1/(2**(beta) *z))*whitw(k, (1/4 + l*(l+1))**(1/2), 2*z*(1+mu)**(1/2))
        w= +w_s0 +w_s
        W.append(w)
        
    W=np.array(W)
    
    return(W)
```
